```python
import torch
import torchaudio
from wav2vec2_ja import WAV2VEC2_ASR_BASE_CV12JA
import IPython.display as ipd
```

```python
bundle = WAV2VEC2_ASR_BASE_CV12JA
model = bundle.get_model()
labels = bundle.get_labels()
```

```python
waveform, sr = torchaudio.load('https://github.com/kaiidams/wav2vec2_ja/blob/main/hiroshiba_normal_001.wav')
waveform = torchaudio.functional.resample(waveform, sr, bundle.sample_rate)
```

```python
ipd.Audio(data=waveform.numpy(), rate=bundle.sample_rate)
```

```python
with torch.inference_mode():
    emission, _ = model(waveform)
    indices = torch.argmax(emission[0], dim=-1)  # [num_seq,]
    indices = torch.unique_consecutive(indices, dim=-1)
    indices = [i for i in indices if i != 0]
    predict = " ".join([labels[i] for i in indices])
```

```python
print(predict)
```
